```python
import os
import matplotlib.pyplot as plt
import json
import matplotlib.image as mpimg

rootdir = './test/'

fig, axs = plt.subplots(3,9, figsize=(40,10))
i1, i2,i3   = 0, 0,0

for ax in axs:
    for a in ax:
        a.set_axis_off()

for subdir, dirs, files in os.walk(rootdir):
    for dir in sorted(dirs):
        if (dir!=".ipynb_checkpoints"):
            print(rootdir+dir+"/experiment.json")
            f = open(rootdir+dir+"/experiment.json")
            data = json.load(f)
            if data["parameters"]["name_model"]=="Prueba":
                img = mpimg.imread(rootdir+dir+"/accuracy.png")
                axs[0,i1].imshow(img)
                axs[0,i1].set_title(str(data["parameters"]["initial_lr"])+" - "+ str(data["test_accuracy"]))
                #axs[1,i1].imread(rootdir+dir+"/loss.png")
                i1+=1
            if data["parameters"]["name_model"]=="MLP-4layers":
                img = mpimg.imread(rootdir+dir+"/accuracy.png")
                axs[1,i2].imshow(img)
                axs[1,i2].set_title(str(data["parameters"]["initial_lr"])+" - "+ str(data["test_accuracy"]))
                #axs[1,i2].imread(rootdir+dir+"/loss.png")
                i2+=1
            
            if data["parameters"]["name_model"]=="MLP-10layers":
                img = mpimg.imread(rootdir+dir+"/accuracy.png")
                axs[2,i3].imshow(img)
                axs[2,i3].set_title(str(data["parameters"]["initial_lr"])+" - "+ str(data["test_accuracy"]))
                #axs[1,i2].imread(rootdir+dir+"/loss.png")
                i3+=1
            
```
